```python
# Import the file mailinglist_unformatted.csv into a pandas dataframe

import pandas as pd

# Read the file into a pandas dataframe
df = pd.read_csv('mailinglist_unformatted.csv', sep='\t')

# Print the dataframe
display(df)

# Remove duplicates and show where they were removed
print('Removing duplicates...')
print(df.duplicated())

# Remove duplicates
df = df.drop_duplicates()

# reset the index
df = df.reset_index(drop=True)

# Print the dataframe
display(df)

# Import email_no_name.csv into a pandas dataframe
no_name_df = pd.read_csv('email_no_name.csv', sep='\t')

# combine the two dataframes
df = pd.concat([df, no_name_df])


display(df)
```

```python
# replace naN values with an empty string
df = df.fillna('')
display(df)

# Save the formatted output to a text file
with open('./formatted_mailinglist.txt', 'w') as f:
    for index, row in df.iterrows():
        f.write(f"{row['Email']} {row['Name']}\n")

# Save the formatted output to a csv file
df.to_csv('./formatted_mailinglist.csv', index=False)
```
